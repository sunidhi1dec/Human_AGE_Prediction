```python
y_col_name = 'BMI'
y = data[y_col_name]
def scatter(x_col_name):
    plt.scatter(df[x_col_name], y)
    plt.xlabel(x_col_name)
    plt.ylabel(y_col_name)
    plt.title('Finding Linear Relation')
    plt.show()
```

```python
def show_linearity_subplots_for(df):
    fig, ax = plt.subplots(1,3)
    fig.set_figheight(5)
    fig.set_figwidth(20)
    fig.subplots_adjust(left=.2, bottom=None, right=None, top=None, wspace=.2, hspace=.2)
    plt1 = plt.subplot(1,3,1)
    plt2 = plt.subplot(1,3,2)
    plt3 = plt.subplot(1,3,3)

    plt1.scatter(df[df.columns[6]], df[df.columns[7]])
    plt1.set_xlabel(df.columns[6])
    plt1.set_ylabel(df.columns[7])
    plt1.grid(axis='both', alpha=.25)

    plt2.scatter(df[df.columns[5]], df[df.columns[6]])
    plt2.set_xlabel(df.columns[5])
    plt2.set_ylabel(df.columns[6])
    plt2.grid(axis='both', alpha=.25)

    plt3.scatter(df[df.columns[1]], df[df.columns[3]])
    plt3.set_xlabel(df.columns[1])
    plt3.set_ylabel(df.columns[3])
    plt3.grid(axis='both', alpha=.25)
```

```python
df_norm = (df-np.mean(df, axis = 0))/np.std(df, axis = 0)
```

```python
df_norm.head()
```

```python
show_linearity_subplots_for(df_norm)
```
